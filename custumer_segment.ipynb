SCALING

````python
x = df.iloc[:,2: ].values
x
````

````python
scale = MinMaxScaler() #This scaler transforms features by scaling them to a given range (by default, [0, 1])
features = scale.fit_transform(x) #fit_transform(x) fits the scaler to the data x and transforms x accordingly This operation scales each feature to the specified range
features # It is a variable contains the scaled data, where each feature has been transformed individually to ensure it falls within the specified range
````

````python
inertia = [] #Inertia measures how well the clusters are defined by computing the sum of squared distances from each data point to its assigned cluster center
for i in range(1,16):
    k_means = KMeans(n_clusters=i)
    k_means = k_means.fit(features)
    inertia.append(k_means.inertia_)
````

````python
# elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16), inertia, 'bo-')
plt.xlabel('No of clusters'), plt.ylabel('Inertia')

# Kelbow visualizer
plt.subplot(1,2,2)
kmeans = KMeans()
visualize = KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()
plt.show()
````

````python
#silhouette measure of how similar an object is to its own cluster (cohesion) compared to other clusters
silhouette_avg = []
for i in range(2,16):
    # initialise kmeans
    kmeans = KMeans(n_clusters= i)
    cluster_labels = kmeans.fit_predict(features)
    # silhouette score
    silhouette_avg.append(silhouette_score(features,cluster_labels))
````

````python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-') #silhouette_avg is assumed to be a list or array containing the average silhouette score for each corresponding number of clusters
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis For Optimal k')
plt.show()
````

K MEAN MODEL

TAKING K VALUE AS 3 PER ELBOW

````python
model = KMeans(n_clusters=3)
model = model.fit(features) # During this process, the model assigns data points to clusters and computes cluster centers (centroids) based on the feature values
````

````python
y_km = model.predict(features)
centers = model.cluster_centers_ # retrieves the coordinates of the centroids (centers) of the clusters identified by the K-means model Each centroid represents the mean coordinates of data points assigned to that cluster
````
